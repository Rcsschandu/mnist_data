```python
df =  pd.read_csv (r'C:\Users\ADMIN\Downloads\mnist_data (1)\df.csv') 
df
```

```python
x = df.iloc[:,0:784]
y = df["label"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=100)
```

## KNN-CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(X_train, y_train)
```

```python
y_test_pred_knn = classifier.predict(X_test)
```

```python
knn_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred_knn})
knn_df.head()
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_test_pred_knn))
```

## Logistic-Regression

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
```

```python
y_test_pred_log = logistic.predict(X_test)
```

```python
knn_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred_log})
knn_df.head()
```

```python
print(accuracy_score(y_test,y_test_pred_log))
```

## Support Vector Machines

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
```

```python
y_test_pred_svc = svc.predict(X_test)
```

```python
svm_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred_svc})
svm_df.head()
```

```python
print(accuracy_score(y_test,y_test_pred_svc))
```

## Decision-Tree

```python
from sklearn.tree import DecisionTreeClassifier
DTclassifier  = DecisionTreeClassifier()
DTclassifier.fit(X_train, y_train)
```

```python
y_test_pred_DTC = DTclassifier.predict(X_test)
```

```python
dtc_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred_DTC})
dtc_df.head()
```

```python
print(accuracy_score(y_test,y_test_pred_DTC))
```

# Random-Forest

```python
from sklearn.ensemble import RandomForestClassifier
RFclassifier = RandomForestClassifier()
RFclassifier.fit(X_train, y_train)
```

```python
y_test_pred_rtc = RFclassifier.predict(X_test)
```

```python
rfc_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred_rtc})
rfc_df.head()
```

```python
print(accuracy_score(y_test,y_test_pred_rtc))
```
